```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('prim_features.csv')

# Mostrar las primeras 10 filas
df.head(10)
```

```python
# Leer el archivo CSV
df2 = pd.read_csv('sec_features.csv')

# Mostrar las primeras 10 filas
df2.head(10)
```

```python
df_total = pd.concat([df, df2], axis=0, ignore_index=True)
```

```python
print(df_total.shape)     # Ver número total de filas y columnas
df_total.head()           # Muestra las primeras filas
df_total.tail()           # Muestra las últimas filas
```

```python
# Contar ocurrencias de cada movieId
conteo_ids = df_total['movieId'].value_counts()

# Filtrar solo los movieId que se repiten
ids_duplicados = conteo_ids[conteo_ids > 1]

# Mostrar resultados
print(f"🔍 Total de movieId duplicados: {len(ids_duplicados)}\n")
print("🆔 MovieIds duplicados y su frecuencia:")
print(ids_duplicados)
```

```python
# Elimina duplicados manteniendo la primera ocurrencia
df_total_sin_duplicados = df_total.drop_duplicates(subset='movieId', keep='first')

# Verifica el resultado
print(f"Número de filas antes: {df_total.shape[0]}")
print(f"Número de filas después: {df_total_sin_duplicados.shape[0]}")
```

```python
# Suponiendo que la primera columna es 'MovieID'
X = df_total_sin_duplicados.iloc[:, 1:]  # Excluye MovieID
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# PCA para reducir a 2 componentes
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X)

# Visualizar
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('PCA - 2 Componentes Principales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
import umap.umap_ as umap
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Aplica PCA para reducción inicial
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X)

# 2. Aplica UMAP para reducción final a 2D
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
X_umap = reducer.fit_transform(X_pca)

# 3. KMeans sobre los datos originales
kmeans = KMeans(n_clusters=19, random_state=0)
labels_kmeans = kmeans.fit_predict(X)

# 4. Visualiza en 2D
plt.figure(figsize=(10, 6))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=labels_kmeans, cmap='Spectral', s=10)
plt.title('K-Means con PCA (50) + UMAP (2D)')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances

# Crear nuevo dataframe con movieId, cluster y UMAP coords
df_umap = pd.DataFrame(X_umap, columns=['UMAP_1', 'UMAP_2'])
df_umap['movieId'] = df_total_sin_duplicados['movieId'].values
df_umap['cluster'] = labels_kmeans

def recomendar_peliculas_umap(movie_id, n_recomendaciones=5):
    if movie_id not in df_umap['movieId'].values:
        print(f"❌ movieId {movie_id} no encontrado.")
        return []

    # Vector y cluster de la película
    pelicula = df_umap[df_umap['movieId'] == movie_id].iloc[0]
    movie_vector = np.array([[pelicula['UMAP_1'], pelicula['UMAP_2']]])
    movie_cluster = pelicula['cluster']

    # Películas en el mismo clúster
    mismo_cluster = df_umap[df_umap['cluster'] == movie_cluster]

    # Distancias en espacio UMAP
    vectors = mismo_cluster[['UMAP_1', 'UMAP_2']].values
    ids = mismo_cluster['movieId'].values
    distancias = euclidean_distances(movie_vector, vectors)[0]

    # Ordenar y obtener los más cercanos (sin incluir la misma)
    indices = distancias.argsort()
    recomendaciones = [ids[i] for i in indices if ids[i] != movie_id][:n_recomendaciones]

    return recomendaciones

```

```python
recomendaciones = recomendar_peliculas_umap(movie_id=96430)
print("Películas recomendadas:", recomendaciones)
```

```python
df_umap = df_umap.merge(df3[['movieId','title', 'poster_url','poster_url_tmbd']], on='movieId', how='left')
```

```python
import requests

def url_valida(url):
    try:
        if not url or pd.isna(url):
            return False
        response = requests.head(url, allow_redirects=True, timeout=5)
        return response.status_code == 200
    except:
        return False

def recomendar_peliculas_umap(movie_id, n_recomendaciones=5):
    if movie_id not in df_umap['movieId'].values:
        print(f"❌ movieId {movie_id} no encontrado.")
        return []

    pelicula = df_umap[df_umap['movieId'] == movie_id].iloc[0]
    movie_vector = np.array([[pelicula['UMAP_1'], pelicula['UMAP_2']]])
    movie_cluster = pelicula['cluster']

    mismo_cluster = df_umap[df_umap['cluster'] == movie_cluster].copy()
    vectors = mismo_cluster[['UMAP_1', 'UMAP_2']].values
    distancias = euclidean_distances(movie_vector, vectors)[0]

    recomendaciones_idx = [i for i in distancias.argsort() if mismo_cluster.iloc[i]['movieId'] != movie_id][:n_recomendaciones]
    recomendaciones = mismo_cluster.iloc[recomendaciones_idx].copy()

    for col in ['title', 'poster_url', 'poster_url_tmbd']:
        if col not in recomendaciones.columns:
            recomendaciones[col] = None

    print(f"\n📽️ Recomendaciones para '{pelicula.get('title', 'Título no disponible')}' (movieId={movie_id}):")
    for _, row in recomendaciones.iterrows():
        poster = None

        # Intentar validar poster_url
        if url_valida(row['poster_url']):
            poster = row['poster_url']
        # Si no, intentar con poster_url_tmbd
        elif url_valida(row['poster_url_tmbd']):
            poster = row['poster_url_tmbd']
        # Si ninguno funciona, mostrar aviso
        else:
            poster = "❌ No disponible"

        print(f"- {row['title']} (ID: {row['movieId']})")
        print(f"  Poster: {poster}")

    return recomendaciones
```

```python
from IPython.display import display, HTML

def recomendar_peliculas_umap_con_imagen_horizontal(movie_id, n_recomendaciones=10):
    if movie_id not in df_umap['movieId'].values:
        print(f"❌ movieId {movie_id} no encontrado.")
        return []

    pelicula = df_umap[df_umap['movieId'] == movie_id].iloc[0]
    movie_vector = np.array([[pelicula['UMAP_1'], pelicula['UMAP_2']]])
    movie_cluster = pelicula['cluster']

    mismo_cluster = df_umap[df_umap['cluster'] == movie_cluster].copy()
    vectors = mismo_cluster[['UMAP_1', 'UMAP_2']].values
    distancias = euclidean_distances(movie_vector, vectors)[0]

    recomendaciones_idx = [i for i in distancias.argsort() if mismo_cluster.iloc[i]['movieId'] != movie_id][:n_recomendaciones]
    recomendaciones = mismo_cluster.iloc[recomendaciones_idx].copy()

    for col in ['title', 'poster_url', 'poster_url_tmbd']:
        if col not in recomendaciones.columns:
            recomendaciones[col] = None

    print(f"\n📽️ Recomendaciones para '{pelicula.get('title', 'Título no disponible')}' (movieId={movie_id}):")

    html_content = "<div style='display:flex; gap:20px; flex-wrap: wrap;'>"
    for _, row in recomendaciones.iterrows():
        poster = row['poster_url']
        if not url_valida(poster):
            poster = row['poster_url_tmbd']
        if poster is None or not url_valida(poster):
            poster = ""  # puedes poner URL de una imagen genérica aquí si quieres

        html_content += f"""
        <div style='flex: 0 0 auto; text-align:center; width: 180px;'>
            <b style="font-size:14px;">{row['title']}</b><br>
            <img src="{poster}" alt="Poster" style="height:250px; width:auto; margin-top:5px;"><br>
            <small>ID: {row['movieId']}</small>
        </div>
        """
    html_content += "</div>"

    display(HTML(html_content))
    return recomendaciones
```
